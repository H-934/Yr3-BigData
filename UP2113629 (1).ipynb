```python
#Exersice 1

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Promts the user to enter a number
number_input = int(input("Enter a number: "))

# Checks if the number is prime or not and prints the result
if is_prime(number_input):
    print("This is a prime number")
else:
    print("This not a prime number")
```

```python
#Exercise 2

def fibonacci(n):
    sequence = [0, 1]
    for _ in range(n - 2):
        sequence.append(sequence[-1] + sequence[-2])
    return sequence[:n]

try:
    num = int(input("Enter how many numbers in Fibonacci sequence to display: "))
    if num < 0:
        print("Please enter a non-negative integer.")
    else:
        print("Fibonacci sequence:", fibonacci(num))
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
#Exercise 3

def print_decreasing_triangle(rows):
    for i in range(rows, 0, -1):
        for j in range(i):
            print("*", end=" ")  
        print() 

#prints the decreasing triangle pattenr
rows = 5
print_decreasing_triangle(rows)
```

```python
#Exercise 6

import random

def guess_the_number():
    secret_number = random.randint(1, 10)
    attempts = 0

    while True:
        try:
            guess = int(input("Guess the number (between 1 and 10): "))
            attempts += 1

            if guess < secret_number:
                print("Too low! Try again.")
            elif guess > secret_number:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number {secret_number} correctly in {attempts} attempts!")
                break  # Exit the loop when the guess is correct

        except ValueError:
            print("Invalid input. Please enter an integer.")

if __name__ == "__main__":
    guess_the_number()

#i added number of attempts but i understand it wasnt stated to do so in the exercise promt
```

```python
#Exercose 8

import math

def calculate_series(x, n):
    total_sum = 0
    for i in range(1, n + 1):
        numerator = x * (5** (2 * i))  # Calculate x * (5^(2*i))
        denominator = i + math.factorial(i + 1)
        total_sum += numerator / denominator
    return total_sum

def main():
    try:
        x = float(input("Enter the value of x: "))
        n = int(input("Enter the number of terms (N): "))

        if n <= 0:
            print("N must be a positive integer.")
            return

        result = calculate_series(x, n)
        print("The sum of the series is:", result)

    except ValueError:
        print("Invalid input. Please enter numeric values for x and N.")

if __name__ == "__main__":
    main()
```

```python
#Exercise 9

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# Load the image from the URL
url = 'https://github.com/H-934/Yr3-BigData/blob/main/CW_PHOTO.jpg?raw=true'
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Convert to NumPy array
image_array = np.array(image)

# Create mirrored versions
horizontal_mirror = np.fliplr(image_array)  # Flip left to right
vertical_mirror = np.flipud(image_array)    # Flip up to down

# Display original and mirrored images
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(image_array)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(horizontal_mirror)
axes[1].set_title("Horizontal Mirror")
axes[1].axis("off")

axes[2].imshow(vertical_mirror)
axes[2].set_title("Vertical Mirror")
axes[2].axis("off")

plt.show()
```

```python
#Exercise 10

import numpy as np

temperatures_celsius = [22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23, 24, 26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23]

# 1) Create a NumPy array
temperatures_np = np.array(temperatures_celsius)

# 2) Find the minimum, maximum, and average temperature
min_temp = np.min(temperatures_np)
max_temp = np.max(temperatures_np)
avg_temp = np.mean(temperatures_np)

print(f"Minimum temperature: {min_temp}C")
print(f"Maximum temperature: {max_temp}C")
print(f"Average temperature: {avg_temp:.2f}C")

# 3) Convert the temperatures from Celsius to Fahrenheit
temperatures_fahrenheit = (temperatures_np * 9/5) + 32
print(f"Temperatures in Fahrenheit: {temperatures_fahrenheit}")

# 4) Find all days where the temperature was above 25C
days_above_25 = temperatures_np[temperatures_np > 25]
print(f"Days above 25C: {days_above_25}")

# 5) Count how many days had temperatures below 20C
days_below_20 = np.sum(temperatures_np < 20)
print(f"Days below 20C: {days_below_20}")

# 6) Sort the temperature data in ascending order
sorted_temperatures = np.sort(temperatures_np)
print(f"Sorted temperatures: {sorted_temperatures}")

# 7) Reshape the 1D array into a 5x6 2D array
reshaped_temperatures = temperatures_np.reshape(5, 6)
print(f"Reshaped temperatures (5x6):\n{reshaped_temperatures}")
```
